```python
import cv2
```

```python
import matplotlib.pyplot as plt 
```

```python
def main():
    
    path = "C:\\Users\\user\\Downloads\\opencv\\misc\\"
    
    imgpath1 =  path + "4.2.01.tiff"
    imgpath2 =  path + "4.2.03.tiff"
    
    img1 = cv2.imread(imgpath1, 1)
    img2 = cv2.imread(imgpath2, 1)
    
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    
    add=img1+img2
    sub=img1-img2
   
    titles = ['Liquid Drop', 'ARUNESH','addition','subtraction']
    images = [img1, img2,add,sub]
    
    for i in range(4):
        plt.subplot(1, 4, i+1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])

    plt.show()  
 
if __name__ == "__main__":
    main()
    
```
