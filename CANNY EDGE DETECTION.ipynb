```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def main():
    imgpath1="C:\\Users\\user\\Downloads\\opencv\\misc\\5.1.11.tiff"
    img1=cv2.imread(imgpath1,1)
    img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    
    l1=cv2.Canny(img1,50,300,L2gradient=False)
    l2=cv2.Canny(img1,100,150,L2gradient=True)
    titles=['original image','l1 norm','l2 norm']
    outputs=[img1,l1,l2]
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(outputs[i],cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
    
if __name__=="__main__":
    main()
```
