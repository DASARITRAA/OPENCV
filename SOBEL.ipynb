```python
import cv2
```

```python
import  matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def main():
    imgpath1="C:\\Users\\user\\Downloads\\opencv\\misc\\5.1.11.tiff"
    img1=cv2.imread(imgpath1,1)
    img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    
    edgesx=cv2.Sobel(img1,-1,dx=1,dy=0,ksize=11,scale=1,delta=0,borderType=cv2.BORDER_DEFAULT)
    edgesy=cv2.Sobel(img1,-1,dx=0,dy=1,ksize=11,scale=1,delta=0,borderType=cv2.BORDER_DEFAULT)
    edges=edgesx+edgesy                
              
                
    titles = ['Original Image', 'x','y','edges']

    outputs = [img1,edgesx,edgesy,edges]
    
    
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(outputs[i],cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
    
if __name__=="__main__":
    main()
```
