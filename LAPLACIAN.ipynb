```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def main():
    imgpath1="C:\\Users\\user\\Downloads\\opencv\\misc\\5.1.11.tiff"
    img1=cv2.imread(imgpath1,1)
    img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    
    edges=cv2.Laplacian(img1,-1,ksize=31,scale=1,delta=0,borderType=cv2.BORDER_DEFAULT)
                    
              
                
    titles = ['Original Image', 'edges']

    outputs = [img1,edges]
    
    
    for i in range(2):
        plt.subplot(1, 2, i+1)
        plt.imshow(outputs[i],cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
    
if __name__=="__main__":
    main()
```
