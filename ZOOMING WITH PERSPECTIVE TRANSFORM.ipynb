```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import cv2
```

```python
def main():
    imgpath1="C:\\Users\\user\\Downloads\\opencv\\misc\\house.tiff"
    
    img1=cv2.imread(imgpath1,1)
    img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    
    rows,columns,channels=img1.shape
    
    point1=np.float32([[0,0],[400,100],[0,400],[400,400]])
    point2=np.float32([[0,0],[300,0],[0,300],[300,300]])
       
    B=cv2.getPerspectiveTransform(point1,point2)
    
    
    print(B)
    
    
    output=cv2.warpPerspective(img1,B,(300,300))
    plt.imshow(output)
    plt.title('TRANSFORMED IMAGE')
    plt.show()
        
    
if __name__=="__main__":
    main()
```
