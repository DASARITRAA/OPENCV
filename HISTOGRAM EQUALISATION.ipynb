```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def main():
    imgpath1="C:\\Users\\user\\Downloads\\opencv\\misc\\4.1.04.tiff"
    img1=cv2.imread(imgpath1,0)
    
  #  img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    
    output1=cv2.equalizeHist(img1)
    
   # clahe=cv2.createCLAHE()
    clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
    output2=clahe.apply(img1)
    
    output=[img1,output1,output2]
    titles=['original image','equalise hist','clahe']
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(output[i],cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
    
if __name__=="__main__":
    main() 
```
