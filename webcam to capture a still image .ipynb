```python
import cv2
import matplotlib.pyplot as plt
```

```python
def main():
    cap=cv2.VideoCapture(0)
    
    if cap.isOpened():
          ret,frame=cap.read()
          print(ret)
          print(frame)
    else: 
          ret=False
            
    img1=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    plt.imshow(img1)
    plt.title('color image RGB')
    plt.xticks([]) and plt.yticks([])
    plt.show()
    
    cap.release()
    
if  __name__=="__main__":
     main()
    
```
